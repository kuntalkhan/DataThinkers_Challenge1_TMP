```python
columns = [
    "Transaction_ID",
    "Customer_ID",
    "Name",
    "Email",
    "Phone",
    "Address",
    "City",
    "State",
    "Zipcode",
    "Country",
    "Age",
    "Gender",
    "Income",
    "Customer_Segment",
    "Date",
    "Year",
    "Month",
    "Time",
    "Total_Purchases",
    "Amount",
    "Total_Amount",
    "Product_Category",
    "Product_Brand",
    "Product_Type",
    "Feedback",
    "Shipping_Method",
    "Payment_Method",
    "Order_Status",
    "products"
]


print(f"Number of rows before cleaning: {len(df)}")

df.dropna(subset=columns, inplace=True)

print(f"Number of rows after cleaning: {len(df)}")
```

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice style for the plots
sns.set_style("whitegrid")

# --- Customer Demographics Analysis ---

print("--- Generating Customer Demographics Plots ---")
# Create a figure to hold our 3 demographic plots
plt.figure(figsize=(18, 5))

# Plot 1: Age Distribution
plt.subplot(1, 3, 1) # (1 row, 3 columns, 1st plot)
sns.histplot(df['Age'], bins=20, kde=True, color='royalblue')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')

# Plot 2: Gender Distribution
plt.subplot(1, 3, 2)
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')

# Plot 3: Income Level Distribution
plt.subplot(1, 3, 3)
sns.countplot(x='Income', data=df, palette='plasma', order=['Low', 'Medium', 'High'])
plt.title('Customer Income Level Distribution')
plt.xlabel('Income Level')
plt.ylabel('Number of Customers')

plt.tight_layout()
plt.savefig('1')
plt.show()


# --- Spending Habits Analysis ---

print("\n--- Generating Spending Habits Plots ---")
plt.figure(figsize=(15, 6))

# Plot 4: Total Amount Spent by Customer Segment
plt.subplot(1, 2, 1)
sns.boxplot(x='Customer_Segment', y='Total_Amount', data=df, palette='magma')
plt.title('Spending Distribution by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Amount Spent per Transaction')

# Plot 5: Total Amount Spent by Income Level
plt.subplot(1, 2, 2)
sns.boxplot(x='Income', y='Total_Amount', data=df, palette='cividis', order=['Low', 'Medium', 'High'])
plt.title('Spending Distribution by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Total Amount Spent per Transaction')

plt.tight_layout()
plt.savefig('2')
plt.show()

print("\n\n📊 Customer analysis visualizations are complete!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice style for the plots
sns.set_style("whitegrid")

# --- Product Performance Analysis ---

print("--- Generating Product Performance Plot ---")
plt.figure(figsize=(12, 7))

# Find the top 10 most purchased product categories
top_10_categories = df['Product_Category'].value_counts().head(10)

# Create the bar plot
sns.barplot(x=top_10_categories.values, y=top_10_categories.index, palette='viridis', orient='h')
plt.title('Top 10 Most Popular Product Categories')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.savefig('3')
plt.show()


# --- Sales Trend Over Time ---

df['Date'] = pd.to_datetime(df['Date'], format='mixed')

print("\n--- Generating Monthly Sales Trend Plot ---")
# Group sales data by month and sum the total amount. 'M' stands for month-end frequency.
monthly_sales = df.groupby(pd.Grouper(key='Date', freq='M'))['Total_Amount'].sum().reset_index()

plt.figure(figsize=(12, 6))

# Create the line plot
sns.lineplot(x='Date', y='Total_Amount', data=monthly_sales, marker='o', color='purple')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('4')
plt.show()

print("\n\n📦 Product and sales analysis visualizations are complete!")
```

```python
# --- Calculate Core Numbers ---

# Total revenue from all transactions
total_revenue = df['Total_Amount'].sum()

# Total number of unique transactions (or orders)
total_orders = df['Transaction_ID'].nunique()

# Total number of unique customers
total_customers = df['Customer_ID'].nunique()



apv = total_revenue / total_orders



purchase_frequency = total_orders / total_customers


# --- Display the Results ---
print("--- Key Performance Indicators ---")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Unique Orders: {total_orders}")
print(f"Total Unique Customers: {total_customers}\n")

print(f"Average Purchase Value (APV): ${apv:,.2f}")
print(f"Purchase Frequency: {purchase_frequency:.2f} orders per customer")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df['Date'] = pd.to_datetime(df['Date'], format='mixed')



monthly_sales = df.groupby(pd.Grouper(key='Date', freq='ME'))['Total_Amount'].sum().reset_index()



monthly_sales['Growth_Rate'] = monthly_sales['Total_Amount'].pct_change() * 100
monthly_sales['Growth_Rate'].fillna(0, inplace=True)



print("--- Monthly Sales and Growth Rate ---")
print(monthly_sales.to_string(formatters={
    'Total_Amount': '${:,.2f}'.format,
    'Growth_Rate': '{:.2f}%'.format
}))



print("\n--- Generating Sales Growth Rate Visualization ---")
plt.figure(figsize=(12, 6))
sns.barplot(x='Date', y='Growth_Rate', data=monthly_sales, palette='coolwarm')
plt.title('Month-over-Month Sales Growth Rate (%)')
plt.xlabel('Month')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
plt.axhline(0, color='black', linewidth=0.8)
plt.savefig('5')
plt.show()
```

```python

customer_purchases = df.groupby('Customer_ID')['Transaction_ID'].nunique()



repeat_customers_count = (customer_purchases > 1).sum()


total_customers_count = df['Customer_ID'].nunique()



repeat_customer_rate = (repeat_customers_count / total_customers_count) * 100



print("--- Customer Retention Analysis ---")
print(f"Total Unique Customers: {total_customers_count}")
print(f"Number of Repeat Customers (made >1 purchase): {repeat_customers_count}")
print(f"\nRepeat Customer Rate: {repeat_customer_rate:.2f}%")
```
